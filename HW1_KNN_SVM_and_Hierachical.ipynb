```python
import tensorflow_datasets as tfds
import cv2
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
import warnings
warnings.filterwarnings('ignore')
```

# 1.Load dataset & Split train/val/test

```python
train_ds, val_ds, test_ds = tfds.load('caltech101', split=['train[:56%]+test[:56%]', 'train[56%:80%]+test[56%:80%]', 'train[80%:]+test[80%:]'])

print("Train ds size: ", len(train_ds))
print("Test ds size: ", len(val_ds))  
print("Valid ds size: ", len(test_ds)) 
```

# 2.Training model

-Transform DataFrame

```python
train = tfds.as_dataframe(train_ds)
val = tfds.as_dataframe(val_ds)
test = tfds.as_dataframe(test_ds)
```

## Calculate Histogram

```python
def calc_Hist(img):
  hist = cv2.calcHist([img], [0], None, [256], [0, 256])
  size = img.shape[0]*img.shape[1]
  hist = hist / size
  return hist
```

```python
def ReadData(ds):
  hist = []
  
  image_arr = ds['image'].to_numpy()
  label_arr = ds['label'].to_numpy()
  
  for img in image_arr:
    hist.append(calc_Hist(img).reshape(-1))
    images = np.array(hist)
    labels = label_arr
  return images, labels
```

```python
X_train, y_train = ReadData(train)
X_val, y_val = ReadData(val)
X_test, y_test = ReadData(test)
```

## Use KNN

```python
for i in range(1, 17):

  print('k =', i, ':')
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train, y_train)

  y_train_pred = knn.predict(X_train)
  print("Accuracy on Train:", accuracy_score(y_train, y_train_pred))

  y_val_pred = knn.predict(X_val)
  print("Accuracy on Val:", accuracy_score(y_val, y_val_pred))

  y_test_pred = knn.predict(X_test)
  print("Accuracy on Test:", accuracy_score(y_test, y_test_pred))

  print(classification_report(y_test, y_test_pred))

  print("----------------------------------------------------------------------------------------")
```

## Use SVM

```python
for i in range(1, 9):

  print('C =', i, ':')
  model = SVC(C = i)
  model.fit(X_train, y_train)
  y_train_pred = model.predict(X_train)
  print("Accuracy on Train:", accuracy_score(y_train, y_train_pred))
  y_val_pred = model.predict(X_val)
  print("Accuracy on Val:", accuracy_score(y_val, y_val_pred))
  y_test_pred = model.predict(X_test)
  print("Accuracy on Test:", accuracy_score(y_test, y_test_pred))

  print(classification_report(y_test, y_test_pred))

  print("----------------------------------------------------------------------------------------")
```

### SVC (C = 1)

```python
print('C = 1 :')
model = SVC(C = 1)
model.fit(X_train, y_train)
y_train_pred = model.predict(X_train)
print("Accuracy on Train:", accuracy_score(y_train, y_train_pred))
y_val_pred = model.predict(X_val)
print("Accuracy on Val:", accuracy_score(y_val, y_val_pred))
y_test_pred = model.predict(X_test)
print("Accuracy on Test:", accuracy_score(y_test, y_test_pred))

print(classification_report(y_test, y_test_pred))
print("Number of support vectors: ", len(model.support_vectors_))
print("----------------------------------------------------------------------------------------")
```

### SVC (1 vs 1)

```python
model_ovo = SVC(decision_function_shape='ovo')
model_ovo.fit(X_train, y_train)
```
